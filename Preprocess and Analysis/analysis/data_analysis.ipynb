Length, width, curb weight, engine size, horsepower and num of cylinder all have a strong positive correlation with price.

```python
# Scatter Plot (Analysis relationship between other related components and price)
fig, ((lengthprice), (widthprice), (horsepowerprice), (curbprice), (enginesizeprice), (numofcynprice)) = plt.subplots(1, 6, figsize = [40,10], sharey = True)

lengthprice.scatter(auto_imports['length'], auto_imports['price'], color = 'red')
lengthprice.set_title('Length and Price')

widthprice.scatter(auto_imports['width'], auto_imports['price'], color = 'orange')
widthprice.set_title('Width and Price')

horsepowerprice.scatter(auto_imports['horsepower'], auto_imports['price'], color = 'yellow')
horsepowerprice.set_title('HorsePower and Price')

curbprice.scatter(auto_imports['curb-weight'], auto_imports['price'], color = 'green')
curbprice.set_title('Curb Weight and Price')

enginesizeprice.scatter(auto_imports['engine-size'], auto_imports['price'], color = 'blue')
enginesizeprice.set_title('Engine Size and Price')

numofcynprice.scatter(auto_imports['num-of-cylinders'], auto_imports['price'], color = 'purple')
numofcynprice.set_title('Num of Cylinder and Price')

fig.suptitle('Scatter Plot for Auto Imports Price')

plt.show()
```

```python
# Box Plot 
fig, ((length), (width), (horsepower), (curbweight), (enginesize), (numcylinders), (price)) = plt.subplots(1, 7, figsize = [35, 10])

fig.suptitle('Box Plots for Auto Imports')

length.boxplot(auto_imports['length'])
length.set_title('Length')

width.boxplot(auto_imports['width'])
width.set_title('Width')

horsepower.boxplot(auto_imports['horsepower'])
horsepower.set_title('Horsepower')

curbweight.boxplot(auto_imports['curb-weight'])
curbweight.set_title('Curb Weight')

enginesize.boxplot(auto_imports['engine-size'])
enginesize.set_title('Engine Size')

numcylinders.boxplot(auto_imports['num-of-cylinders'])
numcylinders.set_title('Num of Cylinders')

price.boxplot(auto_imports['price'])
price.set_title('Price')
```

```python
# Histogram
fig, ((length), (width), (horsepower), (curbweight), (enginesize), (numcylinders), (price)) = plt.subplots(1, 7, figsize = [35, 10])

fig.suptitle('Box Plots for Auto Imports')

length.hist(auto_imports['length'], fill = False)
length.set_title('Length')
length.axvline(stat.mean(auto_imports['length']), color = 'r')

width.hist(auto_imports['width'], fill = False)
width.set_title('Width')
width.axvline(stat.mean(auto_imports['width']), color = 'r')

horsepower.hist(auto_imports['horsepower'], fill = False)
horsepower.set_title('Horsepower')
horsepower.axvline(stat.mean(auto_imports['horsepower']), color = 'r')

curbweight.hist(auto_imports['curb-weight'], fill = False)
curbweight.set_title('Curb Weight')
curbweight.axvline(stat.mean(auto_imports['curb-weight']), color = 'r')

enginesize.hist(auto_imports['engine-size'], fill = False)
enginesize.set_title('Engine Size')
enginesize.axvline(stat.mean(auto_imports['engine-size']), color = 'r')

numcylinders.hist(auto_imports['num-of-cylinders'], fill = False)
numcylinders.set_title('Number of Cylinders')
numcylinders.axvline(stat.mean(auto_imports['num-of-cylinders']), color = 'r')

price.hist(auto_imports['price'], fill = False)
price.set_title('Price')
price.axvline(stat.mean(auto_imports['price']), color = 'r')
```

### Analysis

__INCOMPLETE/INCORRECT DATA__ All the missing data from the raw dataset are filled in and replaced by its mean, this normalized dataset is complete. Looking from the histogram and scatter plot, there isn't incorrect data that should be out of the range.

__ERRORS__ From the scatter plot, box plot and histogram, the data is distributed in a normal range and it shows no error in its attributes. All numbers are in its float numerical form.


__OUTLIERS__
The width, horsepower, engine size, num of cylinders and price has a lot more outliers than expected. It seems reasonable for these attributes to have outliers since some cars are expensive, produces higher power and efficiency than the others in terms of the brand and build. 

__CENTRAL TENDENCY/VARIABILITY/DISPERSION__ 
- The value of the length falls between 170 to 180 that has many values of 170 some and is evenly spread.
- The width is evenly spread. It has outliers of > 70 and is normally distributed. Width is centralized from 65 to 67.
- Horsepower is slightly skewed but not as bad. It is centralized between 95 to 100 with more values less than a 100. It is somewhat evenly spread.
- Curb Weight is centered around 2300 to 2600 but most of its values are between 2400 to 2500. It is evenly spread.
- Engine Size is centered around 80 to 120, most of its data are also distributed within that range. It is skewed with most of its value at the left of the range with some outliers after value of 300.
- Num of Cylinders is badly skewed with most of its value in 1 and many outliers > 1.
- Price is distributed more to the left (more cheaper cars) within a range of 10000 to 15000. 

__RELATIONSHIP WITH PRICE (USING SCATTERPLOT & CORRELATION)__
- Stroke, compression ratio, make, engine type, peak rpm, etc. (see correlation chart) has no correlation with price.
- Horsepower has a positive correlation with price, as price increases, horsepower increases along with length, width, curb weight, engine size and number of cylinders.
- Both city mpg, highway mpg, etc (see correlation chart) has a negative correlation with price, as price decreases, both mpg decreases.

According to the correlations, we will be using length, width, curb weight, engine size and number of cylinders to predict our price.
