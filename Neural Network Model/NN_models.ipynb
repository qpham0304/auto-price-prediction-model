```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn import neighbors
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error 
from math import sqrt
```

#### Load Dataset

```python
# Load data
auto_imports = pd.read_csv("./input/processed_data.csv")

# Remove features with poor coverage or lower relevance and keep price target column
columns_to_drop = ['symboling','normalized-losses','wheel-base','height', 'bore','stroke',
'compression-ratio','peak-rpm','city-mpg','highway-mpg','price','make','fuel-type','aspiration',
'num-of-doors','body-style','drive-wheels','engine-location','engine-type','fuel-system','num-of-cylinders']
    
X_auto = np.c_[auto_imports.drop(columns_to_drop, axis=1)]
y_auto = np.c_[auto_imports['price']]
```

## PART A:

#### Using various number of hidden layers and nodes with different solvers

```python
# 2 hidden layers
X_train, X_test, y_train, y_test = train_test_split(X_auto, y_auto, random_state=0)

for units in [10, 20, 50, 100]:
    for s in ['lbfgs', 'sgd', 'adam']:
        nnclf = MLPRegressor(hidden_layer_sizes = [units, units], solver=s,
                            random_state = 0).fit(X_train, y_train.ravel())

        y_pred = nnclf.predict(X_test)      
            
        print("2 Hidden Layers: ", units, "with Solver: ", s)
        print("RMSE: ", sqrt(mean_squared_error(y_test,y_pred)))
        print("R2 Train Score: ", nnclf.score(X_train, y_train))
        print("R2 Test Score: ", nnclf.score(X_test, y_test))
```

<b>Removing solver sgd (stochastic gradient descent) because it produces infinite value error for the hyperparameters.<b>

```python
# 3 hidden layers
X_train, X_test, y_train, y_test = train_test_split(X_auto, y_auto, random_state=0)

for units in [10, 20, 50, 100]:
    for s in ['lbfgs', 'adam']:
        nnclf = MLPRegressor(hidden_layer_sizes = [units, units, units], solver=s,
                            random_state = 0).fit(X_train, y_train.ravel())

        y_pred = nnclf.predict(X_test)      
            
        print("3 Hidden Layers: ", units, "with Solver: ", s)
        print("RMSE: ", sqrt(mean_squared_error(y_test,y_pred)))
        print("R2 Train Score: ", nnclf.score(X_train, y_train))
        print("R2 Test Score: ", nnclf.score(X_test, y_test))
```

```python
# 4 hidden layers
X_train, X_test, y_train, y_test = train_test_split(X_auto, y_auto, random_state=0)

for units in [10, 20, 50, 100]:
    for s in ['lbfgs', 'adam']:
        nnclf = MLPRegressor(hidden_layer_sizes = [units, units, units, units], solver=s,
                            random_state = 0).fit(X_train, y_train.ravel())

        y_pred = nnclf.predict(X_test)      
            
        print("4 Hidden Layers: ", units, "with Solver: ", s)
        print("RMSE: ", sqrt(mean_squared_error(y_test,y_pred)))
        print("R2 Train Score: ", nnclf.score(X_train, y_train))
        print("R2 Test Score: ", nnclf.score(X_test, y_test))
```

```python
# 5 hidden layers
X_train, X_test, y_train, y_test = train_test_split(X_auto, y_auto, random_state=0)

for units in [10, 20, 50, 100]:
    for s in ['lbfgs', 'adam']:
        nnclf = MLPRegressor(hidden_layer_sizes = [units, units, units, units, units], solver=s,
                            random_state = 0).fit(X_train, y_train.ravel())

        y_pred = nnclf.predict(X_test)      
            
        print("5 Hidden Layers: ", units, "with Solver: ", s)
        print("RMSE: ", sqrt(mean_squared_error(y_test,y_pred)))
        print("R2 Train Score: ", nnclf.score(X_train, y_train))
        print("R2 Test Score: ", nnclf.score(X_test, y_test))
```

***Out of all the number of units with different hidden layers and solver, the model with a hidden layers of 2 with units 50 has the smallest RMSE and highest possible r2 scores for test and train data. The model is optimized when solver is lbfgs.

#### Using different activation functions

```python
for a in ['tanh', 'relu', 'logistic']:
    nnclf = MLPRegressor(hidden_layer_sizes = [50,50], solver='lbfgs',
                        activation=a, random_state = 0).fit(X_train, y_train.ravel())

    y_pred = nnclf.predict(X_test)      
            
    print("\nActivation: ", a)
    print("RMSE: ", sqrt(mean_squared_error(y_test,y_pred)))
    print("R2 Train Score: ", nnclf.score(X_train, y_train))
    print("R2 Test Score: ", nnclf.score(X_test, y_test))
```

***Relu is the best optimal activation function out of the three functions.

#### Using alpha parameter

```python
for alp in [0.01, 0.1, 1.0, 5.0]:
    nnclf = MLPRegressor(hidden_layer_sizes = [50,50], solver='lbfgs',
                        activation='relu', alpha=alp,
                        random_state = 0).fit(X_train, y_train.ravel())

    y_pred = nnclf.predict(X_test)      
            
    print("\nAlpha: ", alp)
    print("RMSE: ", sqrt(mean_squared_error(y_test,y_pred)))
    print("R2 Train Score: ", nnclf.score(X_train, y_train))
    print("R2 Test Score: ", nnclf.score(X_test, y_test))
```

***Alpha = 0.1 is the optimal parameter.

#### Using max iterations

```python
for i in [20, 50, 100, 200, 500]:    
    nnclf = MLPRegressor(hidden_layer_sizes = [50,50], solver='lbfgs',
                        activation='relu', alpha=0.1, max_iter=i,
                        random_state = 0).fit(X_train, y_train.ravel())

    y_pred = nnclf.predict(X_test)      
            
    print("\nMax Iteration: ", i)
    print("RMSE: ", sqrt(mean_squared_error(y_test,y_pred)))
    print("R2 Train Score: ", nnclf.score(X_train, y_train))
    print("R2 Test Score: ", nnclf.score(X_test, y_test))
```

***The optimal max iteration for this model is 500.

Since the solver we are using is lbfgs, there are some parameters that we don't need to use as a parameter.

## PART B:

Model Evaluations

```python
# Scaling features
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

X_train, X_test, y_train, y_test = train_test_split(X_auto, y_auto, test_size=0.1,random_state=0)

# Scaling
X_train_scaled = scaler.fit_transform(X_train)
X_train = pd.DataFrame(X_train_scaled)

X_test_scaled = scaler.transform(X_test)
X_test = pd.DataFrame(X_test_scaled)
```

```python
# Create models
linear_model = LinearRegression().fit(X_train, y_train)
ridge_model = Ridge(alpha = 1, max_iter=10000).fit(X_train, y_train)
knn_model = neighbors.KNeighborsRegressor(n_neighbors=2).fit(X_train, y_train)
nn_models = MLPRegressor(hidden_layer_sizes = [50,50], solver='lbfgs',
                        activation='relu', alpha=0.1, max_iter = 1000, random_state = 0).fit(X_train, y_train.ravel())
```

```python
# Predicted
y_pred_linear = linear_model.predict(X_test)
y_pred_ridge = ridge_model.predict(X_test)
y_pred_knn = knn_model.predict(X_test)
y_pred_nn = nn_models.predict(X_test)
```

```python
# Recording the values of each model
print("\nLinear Model: ")
print("RMSE: ", sqrt(mean_squared_error(y_test,y_pred_linear)))
print("R2 Train Score: ", linear_model.score(X_train, y_train))
print("R2 Test Score: ", linear_model.score(X_test, y_test))

print("\nRidge Model: ")
print("RMSE: ", sqrt(mean_squared_error(y_test,y_pred_ridge)))
print("R2 Train Score: ", ridge_model.score(X_train, y_train))
print("R2 Test Score: ", ridge_model.score(X_test, y_test))

print("\nKNN Model: ")
print("RMSE: ", sqrt(mean_squared_error(y_test,y_pred_knn)))
print("R2 Train Score: ", knn_model.score(X_train, y_train))
print("R2 Test Score: ", knn_model.score(X_test, y_test))

print("\nNeural Network Model: ")
print("RMSE: ", sqrt(mean_squared_error(y_test,y_pred_nn)))
print("R2 Train Score: ", nn_models.score(X_train, y_train))
print("R2 Test Score: ", nn_models.score(X_test, y_test))
```

#### Learning Curves

```python
# Learning Curve for NN
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt


train_sizes = np.linspace(0.2, 1, 50)

train_sizes, train_mse, test_mse = learning_curve(nn_models, X_auto, y_auto.ravel(), 
                                    train_sizes = train_sizes, 
                                    scoring='neg_root_mean_squared_error', 
                                    cv=10, shuffle=True)

train_scores = np.sqrt(np.abs(train_mse))
test_scores = np.sqrt(np.abs(test_mse))

train_scores_mean = np.mean(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)

# Plot learning curve
fig, ((rmse, r2)) = plt.subplots(1, 2, figsize=(25, 5))
    
rmse.set_title('Learning Curve for Auto Imports (Neural Network, RMSE)')
rmse.set_xlabel("Training examples")
rmse.set_ylabel("Score (RMSE)")
rmse.grid()

rmse.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
rmse.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-Validation score")
rmse.legend(loc="best")

train_sizes = np.linspace(0.2, 1, 50)

train_sizes, train_mse, test_mse = learning_curve(nn_models, X_auto, y_auto.ravel(), 
                                    train_sizes = train_sizes, 
                                    scoring='r2', 
                                    cv=10, shuffle=True)

train_scores = np.sqrt(np.abs(train_mse))
test_scores = np.sqrt(np.abs(test_mse))

train_scores_mean = np.mean(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
    
r2.set_title('Learning Curve for Auto Imports (Neural Network, R2)')
r2.set_xlabel("Training examples")
r2.set_ylabel("Score (R2)")
r2.grid()

r2.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
r2.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-Validation score")
r2.legend(loc="best")

plt.show()
```
